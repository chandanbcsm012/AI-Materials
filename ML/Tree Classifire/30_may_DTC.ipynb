```python
import pandas as pd
```

```python
df=pd.read_csv("c:/dataset/classification/data_tree.txt")
df.color=df.color.map({'green':0,'red':1,'yellow':2})
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
dtc=DecisionTreeClassifier(criterion='gini')
dtc.fit(X,y)
plt.figure(figsize=(15,8))
tree.plot_tree(dtc,feature_names=['color','dim'],filled=True,rounded=True)
plt.show()
```

```python
dtc=DecisionTreeClassifier(criterion='entropy')
dtc.fit(X,y)
plt.figure(figsize=(15,8))
tree.plot_tree(dtc,feature_names=['color','dim'],filled=True,rounded=True)
plt.show()
```

```python
df=pd.read_csv('c:/dataset/classification/iris.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
dtc=DecisionTreeClassifier(criterion="gini",max_features=2,max_depth=7,min_samples_split=3,min_samples_leaf=3)
dtc.fit(X,y)
plt.figure(figsize=(15,8))
tree.plot_tree(dtc,feature_names=df.columns,filled=True,rounded=True)
plt.show()
```
