```python
import cv2
cv2.__version__
```

```python
img_path = './deepika-padukone.jpg'
```

```python
img=cv2.imread(img_path,cv2.IMREAD_COLOR)
print(type(img))
print(img.ndim)
print(img.shape)
```

```python
cv2.imshow('color',img) #create window to display image
cv2.waitKey(9000)#max time to disappear this window in ms(this window will also disappear if a key is pressed before this time )
cv2.destroyWindow('color') #destoy window from memory
```

```python
img_clr=cv2.imread(img_path,cv2.IMREAD_COLOR)
img_gray=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
cv2.imshow('color',img_clr)
cv2.imshow('gray',img_gray)
cv2.waitKey(9000)
cv2.destroyWindow('color')
cv2.destroyWindow('gray')
```

```python
img_clr=cv2.imread(img_path,cv2.IMREAD_COLOR)
img_gray=cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
cv2.imshow('color',img_clr)
cv2.imshow('gray',img_gray)
cv2.waitKey(9000)
cv2.destroyAllWindows()
```

```python
img_clr=cv2.imread(img_path,cv2.IMREAD_COLOR)
img_new=cv2.resize(img_clr,(300,250))
cv2.imshow('color',img_new)
cv2.waitKey(9000)
cv2.destroyWindow('color')
```

```python
img_clr=cv2.imread(img_path,cv2.IMREAD_COLOR)
img_gray=cv2.cvtColor(img_clr,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray',img_gray)
cv2.waitKey(0)#window will disappear only if a key is pressed
cv2.destroyAllWindows()
```

```python
img_clr=cv2.imread(img_path,cv2.IMREAD_COLOR)
cv2.line(img_clr,(100,100),(300,100),(0,0,255),4)
cv2.putText(img_clr,'Sachin',(100,80),cv2.FONT_HERSHEY_PLAIN,3,(255,0,0),2)
cv2.rectangle(img_clr,(200,110),(350,360),(0,255,0),2)
cv2.imshow('color',img_clr)
cv2.imwrite('sachin.png',img_clr)
cv2.waitKey(0)#window will disappear only if a key is pressed
cv2.destroyAllWindows()
```
