```python
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer 
sw=stopwords.words('english')
sw.remove('not')
sw.remove('no')
sw.remove("don't")
sw.remove('do')
sw.remove("didn't")
sw.remove('did')
ps=PorterStemmer()
```

```python
def clean_text(docs):
    text=re.sub('[^a-zA-Z]'," ",docs)
    afterlower=text.lower()
    words=word_tokenize(afterlower)
    nw=[]
    for w in words:
        if(w not in sw):
            ps.stem(w)
            nw.append(w)
    return " ".join(nw)
# load dataset
df=pd.read_csv('../../dataset/sentiment/twitter.csv',nrows=7000,encoding='latin')
df.drop(['ItemID'],axis=1,inplace=True)
df['SentimentText']=df['SentimentText'].astype(str)
df['SentimentText']=df['SentimentText'].map(clean_text)
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,1))
X=cv.fit_transform(df['SentimentText']).toarray()
y=df['Sentiment']
from sklearn.naive_bayes import MultinomialNB
```

```python
mn=MultinomialNB(alpha=0.01)
mn.fit(X,y)
print("Score:",mn.score(X,y))
```
