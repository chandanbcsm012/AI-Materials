```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('../dataset/regression/salary_1_variable.csv')
df.shape
X=df.iloc[:,0:-1].values
y=df.iloc[:,-1].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
pred_train=lr.predict(X_train)
pred_test=lr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_train.flatten(),y_train)#Actual data points
plt.plot(X_train.flatten(),pred_train,c='r')#prediction as line
plt.show()
```

```python
r2_score(y_train,pred_train)
```

```python
plt.scatter(X_test.flatten(),y_test)#Actual data points
plt.plot(X_test.flatten(),pred_test,c='r')#prediction as line
plt.show()
```

```python
r2_score(y_test,pred_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_train,pred_train)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,pred_test)
```
