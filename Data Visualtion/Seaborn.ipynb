# Seaborn

```python
import seaborn as sbn
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/analysis_prepro/tips.csv')
df.head()
```

```python
sbn.countplot(x='gender',data= df)
plt.show()
```

```python
%matplotlib inline
```

```python
sbn.countplot(x='day',data= df, hue='gender')
plt.show()
```

```python
sbn.countplot(x='day',data= df, hue='smoker')
plt.show()
```

```python
sbn.scatterplot(x='total_bill', y='tip', data=df, hue='gender')
plt.show()
```

```python
df_fruit = pd.read_csv('../dataset/classification/fruits.csv')
sbn.scatterplot(x='diameter', y='weight', hue='FruitName', data=df_fruit)
plt.show()
```

```python
df_fruit = pd.read_csv('../dataset/classification/fruits.csv')
sbn.scatterplot(x='diameter', y='weight', hue='FruitName', size='FruitName', data=df_fruit)
plt.show()
```

```python
sbn.distplot(a=df.total_bill, bins=10, hist=False, hist_kws={'edgecolor':'k', 'linewidht':1, "color":'r'})
```

```python
sbn.lineplot(x='total_bill', y='tip', data=df.head(30))
plt.show()
```
