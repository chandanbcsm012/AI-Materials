# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Line Plot
> it used for understanding of two valriable

```python
x = [2,4,6,8,10]
y = [10,12,14,15,16]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('python average salary growth')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.show()
```

```python
plt.plot(x,y, '-r')
plt.title('average salary growth')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.show()
```

```python
plt.plot(x,y)
plt.title('python average salary growth')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.xticks(range(0,15, 2))
plt.show()
```

```python
plt.plot(x,y)
plt.title('python average salary growth')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.xticks(range(0,15, 2))
plt.yticks([10,11,12,13,14,15,16])
plt.show()
```

```python
plt.plot(x,y)
plt.title('python average salary growth')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.axis([10,15,10,15])
plt.show()
```

```python
plt.plot(x,y)
plt.title('python average salary growth')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.axis([10,15,10,15])

plt.grid()
plt.show()
```

```python
plt.plot(x,y, '-r', ls='--', lw=2)
plt.grid()
plt.show()
```

```python
plt.plot(x,y, 'k', ls='--', lw=2)
plt.grid()
plt.show()
```

```python
plt.plot(x,y, '-r', ls='--', lw=2, marker='o')
plt.grid()
plt.show()
```

```python
plt.plot(x,y, '-r', ls='--', lw=2, marker='<')
plt.grid()
plt.show()
```

```python
plt.plot(x,y, '-r', ls='--', lw=2, marker='o', markerfacecolor='b')
plt.grid()
plt.show()
```

```python
plt.plot(x,y, '-r', ls='--', lw=2, marker='o', markerfacecolor='b', markeredgecolor='k')
plt.grid()
plt.show()
```

```python
plt.plot(x,y, '-r', ls='--', lw=2, marker='o', markerfacecolor='b', markeredgecolor='k', markersize=10 )
plt.grid()
plt.show()
```

```python
plt.plot(x,y, '-r', ls='--', lw=2, marker='o', markerfacecolor='b', markeredgecolor='k', markersize=10 )
plt.grid()
plt.show()
```

# Histogram Plot

> it used to understanding data frequency of given range

```python
df = pd.read_csv('../dataset/analysis_prepro/titanic.csv')
df.head()
```

```python
plt.hist(x=df['Age'])
plt.show()
```

```python
plt.hist(x=df['Age'], bins=20)
plt.show()
```

```python
plt.hist(x=df['Age'], bins=range(0,100, 10), edgecolor='k')
plt.show()
```

# Bar Plot

> it used for single category data frequency

ex: match of bhumrah runs

```python
plt.bar(['bhumrha', 'jadeja'], [5,3])
plt.show()
```

```python
plt.bar(df['gender'].unique(), df['gender'].value_counts())
plt.show()
```

```python
plt.bar(df['gender'].unique(), df['gender'].value_counts(), width=.5)
plt.show()
```

```python
plt.bar(df['gender'].unique(), df['gender'].value_counts(), width=.5, edgecolor='k', lw=5, color=['r', 'b'])
plt.show()
```

```python
plt.pie([15, 30, 45, 10],explode = (0, 0.1, 0, 0), labels = ['Frogs', 'Hogs', 'Dogs', 'Logs'])
plt.show()
```

```python
plt.pie(x=[50,15,5], labels=['AAP','BJP', 'Cong'])
plt.show()
```

```python
plt.pie(x=[50,15,5], labels=['AAP','BJP', 'Cong'], wedgeprops={'linewidth':1, 'edgecolor':'k'}, autopct='%f')
plt.show()
```

```python
plt.pie(x=[50,15,5], labels=['AAP','BJP', 'Cong'], wedgeprops={'linewidth':1, 'edgecolor':'k'}, autopct='%1.2f%%')
plt.show()
```

```python
plt.pie(x=[50,15,5], labels=['AAP','BJP', 'Cong'], explode=(0,0,.5), shadow=True, wedgeprops={'linewidth':1, 'edgecolor':'k'}, autopct='%1.2f%%')
plt.savefig('seats_pie.png')
```
