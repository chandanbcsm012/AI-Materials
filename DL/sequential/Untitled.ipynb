```python
import pandas as pd
from tensorflow.keras import Model, Sequential
from tensorflow.keras.layers import Dense,Dropout, Input
```

```python
import tensorflow as tf
(X, y), (Xt, yt) = tf.keras.datasets.mnist.load_data()
(XX, yy), (XXt, YYt) = tf.keras.datasets.fashion_mnist.load_data()
X, Xt, XX, XXt = X/255.0, Xt/255.0, XX/255.0, XXt/255.0
X.shape, y.shape
```

```python
from tensorflow import keras
model = Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(500, activation='relu'),
    keras.layers.Dense(250, activation='relu'),
    keras.layers.Dense(125, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(XX,yy,epochs=10)
```

```python
model.evaluate(XXt,YYt)
```

```python
predictions = model.predict(XXt)
```

```python
import numpy as np
```

```python
for i in range(5000):
    pred, test_org = np.argmax(predictions[i]), yt[i]
    if pred != test_org:
        print(f'{i} === {pred}, {test_org}')
```
