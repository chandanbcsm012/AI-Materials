```python
path = './dataset/docx_files/'
files = os.listdir(path)
key = input("Enter keyword:")
output = []
res = 0
count = 0
for f in files:
    if f.endswith('.docx'):
        count +=1
        data = docxpy.process(path+f)
        if key in data.lower():
            res +=1
            output.append(f)

print("Total files", len(files))
print("Docx files",count)
print("match fles", res)
```
