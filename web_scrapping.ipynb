# Web Scrapping

```python
import requests
```

```python
res = requests.get("https://www.ducatindia.com/")
```

```python
res.text  # return data of the web page as a string
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(res.text, 'html.parser')
only_data = soup.getText(strip=True, separator='\n')
print(only_data)
```

```python
file = open('ducat_index_data.text', 'w')
file.write(only_data)
```

### Get Title of website
> find function return first matching tag

```python
title = soup.find('title')
title.get_text()
```

```python
soup.find('a')
```

```python
all_a = soup.find_all('a')
for a in all_a:
    print(a.get('href'))
```

```python
imgs = soup.find_all('img')
imgs
```

```python
import os
for img in imgs:
    src = img.get('src')
    if src.startswith('http'):
        pass
    elif src.startswith('/'):
        src = 'https://www.ducatindia.com'+src
    r = requests.get(src)
    img_bytes = r.content
    if not os.path.exists('ducat'):
        os.mkdir('ducat')
    name = src[src.rindex('/')+1:]
    file = open('./ducat/'+name, 'wb')
    file.write(img_bytes)
    file.close
print('all saved')
```
